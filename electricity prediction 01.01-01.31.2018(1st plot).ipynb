```python
def prediction(day,month,hour): #in 2018
    
    import csv
    f=open('RealTimeConsumption_2016_2017.csv')
    doc=csv.reader(f)
    str_date= day+'.'+month
    str_time= str(hour)
    lst=[]
    for row in doc:
        if str(row[0][0:5]) == str_date and str(row[1]) == str_time:
            usage =float(int(row[2][0]+row[2][1]+row[2][3]+row[2][4]+row[2][5]+row[2][7]+row[2][8])/100.0)
            lst += [usage]    
    return (lst[0]+lst[1])/2.0
```

```python
time = []
predict = []

for d in range(1,32):
    if len(str(d)) == 1:
        str_d = '0'+str(d)
    else:
        str_d=str(d)
    for h in range(0,24):
        if len(str(h))==1:
            str_h='0'+str(h)+':00'
        else:
            str_h=str(h)+':00'
        time += ['01' + '.' + str_d +'.'+'2018' + ',' + str_h]
        predict += [prediction(str_d,'01',str_h)]

import matplotlib.pyplot as plt
plt.plot(time,predict)
plt.show() #this graphic will show the predictions for Jan 1-31 in 2018
```

```python
import csv
f=open('RealTimeConsumption_01012018_31012018.csv')
doc=csv.reader(f)

time = []
consump = []

for row in doc:
    time += [str(row[0])+','+str(row[1])]
    consump += [str(row[2])]
time.pop(0)
consump.pop(0)


import matplotlib.pyplot as plt
plt.plot(time,consump)
plt.show() #this graphic will show the real time consumption for Jan 1-31 in 2018
```
